```python
from google.colab.patches import cv2_imshow
import cv2
from skimage import io
import numpy as np

try:
    from PIL import Image
except ImportError:
    import Image




for i in range(0,len(reduced_size)):

  # Open the image
  image = Image.open(reduced_size[i])
  
  # Get the original dimensions of the image
  original_width, original_height = image.size
  
  # Determine the ratio to be a max of 1024
  ratio = 1024/original_width

  # Calculate the new dimensions of the image
  new_width = int(original_width * ratio)
  new_height = int(original_height * ratio)
  
  # Resize the image to the new dimensions
  resized_image = image.resize((new_width, new_height))
  
  # Save the resized image
  resized_image.save('/content/drive/My Drive/Thesis/All_Photos/Original_Size_Reduction/Original_TEST/' + str(reduced_size[i][0:3]) + "_disp_Reduced_size" + ".jpg")


  # # new_img = Image.blend(background, overlay, 0.5)
  # new_img.save('/content/drive/My Drive/Thesis/All_Photos/6_ch_try/6_TEST/' + str(i) + "_disp" + ".png","PNG")
```
