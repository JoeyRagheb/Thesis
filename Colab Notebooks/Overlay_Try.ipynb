```python
original = []
disp = []
local_download_path = '/content/drive/My Drive/Thesis/All_Photos/6_ch_try/6_TEST/Separate/'

for filename in os.listdir(local_download_path):
    if '.png' in filename:
        original.append(filename)
original.sort()
print(original)

for filename in os.listdir(local_download_path):
    if '.jpg' in filename: 
        disp.append(filename)
disp.sort()
disparity = disp
print(disp)
```

```python
from google.colab.patches import cv2_imshow
import cv2
from skimage import io
import numpy as np

try:
    from PIL import Image
except ImportError:
    import Image




for i in range(0,len(disparity)):

  background = Image.open(disparity[i])
  overlay = Image.open(original[i])

  background = background.convert("RGBA")
  overlay = overlay.convert("RGBA")

  new_img = Image.blend(background, overlay, 0.5)
  new_img.save('/content/drive/My Drive/Thesis/All_Photos/6_ch_try/6_TEST/' + str(i) + "_disp" + ".png","PNG")
```
