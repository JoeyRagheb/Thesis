```python
plt.rcParams.update({'font.size': 14})

columns = [ '1loss', '7loss']
df = pd.read_csv("Models_accuracy_data.csv", usecols=columns)
one = df['1loss']
one_avg = one.rolling(window=35).mean()
seven = df['7loss']
seven_avg = seven.rolling(window=35).mean()

plt.figure(figsize=(10, 7))
plt.plot(one_avg, 'm-', label='1-Original-photos')
plt.plot(seven_avg, 'k-', label='7-Original-photos-dimension-reduction')
plt.ylabel('Loss')
plt.xlabel('Epochs')
plt.legend(loc='upper right')
plt.show()
```

```python
plt.rcParams.update({'font.size': 16})
columns = [ '1loss', '2loss', '3loss', '4loss', '5loss', '6loss', '7loss']
df = pd.read_csv("Models_accuracy_data.csv", usecols=columns)
# one = df['1loss']
# one_avg = one.rolling(window=1).mean()
two = df['2loss']
two_avg = two.rolling(window=1).mean()
three = df['3loss']
three_avg = three.rolling(window=1).mean()
four = df['4loss']
four_avg = four.rolling(window=1).mean()
five = df['5loss']
five_avg = five.rolling(window=1).mean()
six = df['6loss']
six_avg = six.rolling(window=1).mean()
seven = df['7loss']
seven_avg = seven.rolling(window=1).mean()


plt.figure(figsize=(18.5, 10.5))
# plt.plot(one_avg, 'b-', label='1loss')
plt.plot(two_avg, 'y-', label='2-640-DIFFNet')
plt.plot(three_avg, 'g-', label='3-1024-DIFFNet')
plt.plot(four_avg, 'r-', label='4-RGB-addition')
plt.plot(five_avg, 'm-', label='5-4channels')
plt.plot(six_avg, 'c-', label='6-50-50-overlay')
plt.plot(seven_avg, 'k-', label='7-benchmark-dimension-reduction')
plt.ylabel('Loss')
plt.xlabel('Epochs')
plt.legend(loc='upper right')
plt.show()

# df.plot()
# fig = matplotlib.pyplot.gcf()
# fig.set_size_inches(18.5, 10.5)
# plt.show()
```
