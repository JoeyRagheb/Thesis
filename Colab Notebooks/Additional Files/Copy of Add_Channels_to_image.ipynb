```python
original = []
disp = []
local_download_path = '/content/drive/My Drive/Thesis/All_Photos/Combined/Combined_TRAIN/4_Combine/'

for filename in os.listdir(local_download_path):
    # Original: if 'disp' not in filename and 'npy' not in filename and 'rchive' not in filename:
    if '.jpeg' in filename:
      # Original: original.append(filename[:-4]+'_'+'.jpeg')
        original.append(filename)
original.sort()

'''remove for original
for i in range(0,len(original)):
  original[i] = original[i][:-4]+'.png'
'''
print(original)

for filename in os.listdir(local_download_path):
    # Original: if '.png' in filename:
    if '.png' in filename: 
        disp.append(filename)
disp.sort()
disparity = disp
print(disp)
```

```python
from google.colab.patches import cv2_imshow
import cv2
from skimage import io

# image = cv2.imread('0_disp.png')
# b = image.copy()

for i in range(0,len(disparity)):

  image = cv2.imread(disparity[i]) #'0_disp.png'
  print(disparity[i])
  b = image.copy()
  # set green and red channels to 0
  b[:, :, 1] = 0
  b[:, :, 2] = 0

  g = image.copy()
  # set blue and red channels to 0
  g[:, :, 0] = 0
  g[:, :, 2] = 0

  r = image.copy()
  # set blue and green channels to 0
  r[:, :, 0] = 0
  r[:, :, 1] = 0

  image2 = cv2.imread(original[i])

  b2 = image2.copy()
  # set green and red channels to 0
  b2[:, :, 1] = 0
  b2[:, :, 2] = 0

  g2 = image2.copy()
  # set blue and red channels to 0
  g2[:, :, 0] = 0
  g2[:, :, 2] = 0

  r2 = image2.copy()
  # set blue and green channels to 0
  r2[:, :, 0] = 0
  r2[:, :, 1] = 0

  # cv2_imshow(image)
  # cv2_imshow(image2)
  b3 = b2 + b
  g3 = g2 + g
  r3 = r2 + r

  merged = b3.copy()
  merged[:,:,1] = g3[:,:,1]
  merged[:,:,2] = r3[:,:,2]
  # cv2_imshow(merged)
  cv2.imwrite('/content/drive/My Drive/Thesis/All_Photos/Combined/Combined_TRAIN/4_Combine/Merged' + original[i][:-4] + '_merged.png',merged)
  # cv2.imwrite('/content/drive/My Drive/Colab Notebooks/Thesis/image_and_disparity/Train_Merged/Archive/Merged/' + original[i][:-4] + '_merged.png',merged)
```
